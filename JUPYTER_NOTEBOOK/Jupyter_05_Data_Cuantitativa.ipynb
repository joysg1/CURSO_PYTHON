```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
listings = pd.read_csv("listings.csv")
```

```python
listings
```

```python
plt.hist(listings['price'], bins = np.arange(0,1100,40)) #Determinar la distribuciÃ³n cuantitativa de los precios (histograma)
plt.xlabel('price(in US Dollars)')
plt.show()

```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews']) #Determinar la distribuciÃ³n cuantitativa de los precios en funciÃ³n de los reviews
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews']) #Restringimos el eje x a 1100 con xlim
plt.xlabel('price')
plt.xlim(0,1100)
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'], s=5) #Bajamos el tamaÃ±o de los punto con s
plt.xlabel('price')
plt.xlim(0,1100)
plt.ylabel('number of reviews')
plt.show()
```
