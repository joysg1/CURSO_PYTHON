```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
listings = pd.read_csv("listings.csv")
```

```python
listings
```

```python
plt.hist(listings['price'], bins = np.arange(0,1100,40)) #Determinar la distribuciÃ³n cuantitativa de los precios (histograma)
plt.xlabel('price(in US Dollars)')
plt.show()

```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews']) #Determinar la distribuciÃ³n cuantitativa de los precios en funciÃ³n de los reviews
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews']) #Restringimos el eje x a 1100 con xlim
plt.xlabel('price')
plt.xlim(0,1100)
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'], s=5) #Bajamos el tamaÃ±o de los punto con s
plt.xlabel('price')
plt.xlim(0,1100)
plt.ylabel('number of reviews')
plt.show()
```

```python
import pandas as pd    #Tipos de alojamiento por barrio (top 10)
import matplotlib.pyplot as plt

# Agrupa los datos por barrio y tipo de alojamiento
df_grupo = listings.groupby(['neighbourhood', 'room_type']).size().unstack(fill_value=0)

# Sumar las cantidades por barrio y filtrar los 10 barrios mÃ¡s representativos
top_neighbourhoods = df_grupo.sum(axis=1).nlargest(10).index
df_top = df_grupo.loc[top_neighbourhoods]

# Genera el grÃ¡fico de barras horizontales
df_top.plot(kind='barh', stacked=True, figsize=(10, 6))
plt.title('Tipos de alojamiento por barrio (Top 10)')
plt.xlabel('Cantidad')
plt.ylabel('Barrio')
plt.legend(title='Tipo de alojamiento', bbox_to_anchor=(1, 1))
plt.show()
```
