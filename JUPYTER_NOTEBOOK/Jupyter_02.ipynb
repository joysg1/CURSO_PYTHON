```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del modelo
a = 0.5  # Tasa de reproducción
b = 0.1  # Tasa de mortalidad
N0 = 10  # Población inicial
t_max = 100  # Tiempo máximo de simulación (en meses)

# Crear un arreglo de tiempos
t = np.arange(0, t_max, 1)

# Crear un arreglo de poblaciones
N = np.zeros(len(t))
N[0] = N0

# Simular la dinámica de la población
for i in range(1, len(t)):
  N[i] = N[i-1] + a * N[i-1] - b * N[i-1]

# Crear un gráfico de la población en función del tiempo
plt.plot(t, N)
plt.title('Tasa de reproducción de los conejos')
plt.xlabel('Tiempo (meses)')
plt.ylabel('Población')
plt.show()
```
