```python
#Seccion para los graficos

import matplotlib.pyplot as plt
import seaborn as sns

# GrÃ¡fico de distribuciÃ³n de delitos por dÃ­a de la semana
sns.countplot(x='DAY_OF_WEEK', data=crime)
plt.title('DistribuciÃ³n de delitos por dÃ­a de la semana')
plt.show()
```

```python
# GrÃ¡fico de distribuciÃ³n de delitos por distrito
sns.countplot(x='DISTRICT', data=crime)
plt.title('DistribuciÃ³n de delitos por distrito')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Suponiendo que 'crime' es tu DataFrame
# Obtener las frecuencias de los OFFENSE_CODE_GROUP
frecuencias = crime['OFFENSE_CODE_GROUP'].value_counts()

# Obtener los 5 mÃ¡s frecuentes
top_5 = frecuencias.nlargest(5)

# Definir una lista de colores distintos
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

# Crear un grÃ¡fico de barras horizontales
plt.figure(figsize=(10,6))

for i, (index, value) in enumerate(top_5.items()):
    plt.barh(i, value, color=colors[i])

plt.xlabel('Frecuencia')
plt.ylabel('OFFENSE_CODE_GROUP')
plt.title('Los 5 OFFENSE_CODE_GROUP mÃ¡s frecuentes')
plt.yticks(np.arange(len(top_5)), top_5.index)
plt.xticks(rotation=45)
plt.tight_layout()

# Agregar labels a cada barra
for i, value in enumerate(top_5.values):
    plt.text(value, i, str(value), ha='left', va='center')

plt.show()

```
